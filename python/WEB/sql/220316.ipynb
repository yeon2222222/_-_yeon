```python
sql_16="""
        select ENAME,
        CASE
            WHEN JOB="ANALYST" THEN "분석가"
            WHEN JOB="CLERK" THEN "점원"
            WHEN JOB="MANAGER" THEN "관리자"
            WHEN JOB="PRESIDENT" THEN "사장"
            WHEN JOB="SALESMAN" THEN "판매사원" 
            END AS JOB
        FROM emp
        """
result_16 = _dt.executeAll(sql_16)
df_16=pd.DataFrame(result_16)
df_16
#CASE문 케이스별로 조건지정해줄수 있음->여러 조건일때 편리  #WHEN JOB="SALESMAN" THEN "판매사원" ->ELSE "판매사원"
#IF문 조건의 참거짓만 ->조건이 하나일떄 편리       
```

1. emp table과 dept table을 left join 조건 DEPTNO가 같은 경우
2. 출력 컬럼은 ENAME ,JOB, SAL, DEPTNO
3. ENAME LOC 그대로 출력
4. JOB은 각각 한글명으로 출력
5. SAL은 2450을 기준으로 2450 이상이면 HIGH 미만이면 LOW
6. DEPTNO의 DEPTNO의 값과 dept table의 DNAME값을 더해서 표시 (DEPTNO-DNAME형식)
7.모든 컬럼의 이름을  한글로 변경

```python
sql_17="""
        SELECT 
        emp.ENAME AS "사원명",
        CASE
            WHEN emp.JOB="ANALYST" THEN "분석가"
            WHEN emp.JOB="CLERK" THEN "점원"
            WHEN emp.JOB="MANAGER" THEN "관리자"
            WHEN emp.JOB="PRESIDENT" THEN "사장"
            ELSE "판매사원" 
        END AS "직급",
        IF (emp.SAL>=2450,'HIGH','LOW') AS "월급",
        CONCAT(emp.DEPTNO,"-",dept.DNAME) AS "부서번호-부서명",
        dept.LOC AS "위치"
        FROM emp LEFT JOIN dept 
        ON emp.DEPTNO=dept.DEPTNO
        ORDER BY '사원명' ASC 
        """# 컬럼을 바꾸기전에 사원명으로 데이터를 정렬해서 해줌
result_17 = _dt.executeAll(sql_17)
pd.DataFrame(result_17)
#어떤 테이블에 있는지 구별을 해주는데 컬럼의 이름이 중첩되는 애들은 어느 테이블에 있는 컬럼값인지 정확히 적어줘야함 
#따라서 JOB같은 경우는 구분안해줘도 되지만 DEPTNO는 구분해줘야함->물론 다 적어주는게 좋음
#sql은 FROM뒤부터 실행함.
```

### 부서의 이름이 "sales"인 소속 사원들의 리스트를 출력하라

```python
#사원 이름 emp출력할것
sql_20="""
            SELECT
            ENAME 
            from emp
            LEFT JOIN dept
            ON emp.DEPTNO=dept.DEPTNO
            WHERE dept.DNAME="sales"
        """
result_20=_dt.executeAll(sql_20)
pd.DataFrame(result_20)
```

```python
#다중쿼리문 사용하기
sql_21="""
            SELECT
            ENAME 
            from emp
            WHERE DEPTNO =
            (
                SELECT DEPTNO FROM dept
                WHERE DNAME="sales"
            )
        """
result_21=_dt.executeAll(sql_21)
pd.DataFrame(result_21)
```

### 부서의 이름이 "sales"가 아닌 소속 사원들의 리스트를 출력하라
### 부서의 위치가 'NEWYORK','DALLAS'가 아닌 소속사원들의 리스트를 출력하라.
>NOT 쓰지 않고

```python
sql_22="""
            SELECT
            ENAME 
            from emp
            WHERE DEPTNO IN
            (
                SELECT DEPTNO FROM dept
                WHERE DNAME NOT IN ('sales')
            )
        """
result_22=_dt.executeAll(sql_22)
pd.DataFrame(result_22)
```
