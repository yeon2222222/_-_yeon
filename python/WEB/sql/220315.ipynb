```python
import mod_sql
import pandas as pd
```

```python
_db =mod_sql.Database()
```

```python
sql = """
        select * from sales
        """
result =_db.executeAll(sql)
dt= pd.DataFrame(result)

dt.head()
```

```python
sql_2 = """
            select 
            Country,
            SUM(`Units Sold`) AS 'units sold' 
            from sales 
            group by Country
            order by 'units sold'
        """# 따옴표말고 ``로 묶기
result_2=_db.executeAll(sql_2)
dt_2=pd.DataFrame(result_2)

dt_2.head()
```

```python
sql_2 = """
            select 
            Country,
            avg(`Units Sold`) AS 'units sold' 
            from sales 
            group by Country
            order by `Country` desc
        """# 따옴표말고 ``로 묶기
result_2=_db.executeAll(sql_2)
dt_2=pd.DataFrame(result_2)

dt_2.head()
```

```python
sql_3 = """
            select 
            Country,
            count(`country`) AS cnt 
            from sales 
            group by Country
            order by cnt 
        """# 따옴표말고 ``로 묶기
result_3=_db.executeAll(sql_3)
dt_3=pd.DataFrame(result_3)

dt_3.head()
```
