```python
import pandas as pd

class Change_dt():
    def __init__(self,input_url):#url값만 받아오는 함수
        self.url=input_url
        
    def csv_read(self): #url읽기 -> 매개변수 필요없음
        self.dt=pd.read_csv(self.url) #변수 만들기
        self.dt.sort_values('Country',inplace=True) # 변수를 새로 지정하거나 inplace: 
        self.dt.reset_index(drop=True,inplace=True)
        return self.dt
    def remove_columns(self,input_list=[]):
        self.dt.drop(input_list, axis=1,inplace=True)#input_list는 매개변수이므로 self 안써도됨 self는 이 클래스가 가진 변수
        return self.dt

    def remove_2(self,input_s_column,input_e_column):#시작 끝 콜롬 받아오기 매개변수
        self.dt.drop(self.dt.loc[:,input_s_column:input_e_column],axis=1,inplace=True)
        return self.dt
    
dt= Change_dt("./csv/Sales Records.csv")
```
