```python
import pandas as pd
```

```python
sales_df = pd.read_csv('./Sales Records.csv')
sales_df.head()
```

```python
# 컬럼삭제 ship date,order id, sales channel ,total revenue --원본데이터 유지 (새로운 변수)
# 컬럼명 수정["지역","국가","상품 타입","배송순위","주문날짜","판매개수","단가","원가","총비용","총이윤"] ->원본데이터 유지 x(새로운 변수 유지할 필요없이 계속 같은 변수 애서 작업  )
#country  기준으로 sort
#index값 초기화
```

```python
#컬럼 삭제 drop 축은 열
sales_df_2= sales_df.drop(["Ship Date","Order ID","Sales Channel","Total Revenue"],axis=1)
sales_df_2.head()
```

```python
#컬럼명 변경
sales_df_2.columns=["지역","국가","상품 타입","배송순위","주문날짜","판매개수","단가","원가","총비용","총이윤"]
sales_df_2.head()
```

```python
sales_df_2.sort_values("국가",inplace=True) #오름차순
sales_df_2
```

```python
sales_df_2.reset_index(drop=True,inplace=True)#기존의 인덱스 살리고 싶다 ->drop 안쓰기 , 죽이고 싶다 ->drop=True  
sales_df_2.head()                             # inplace그리고 새로운데이터로 변경
```

```python
#뒤의 소수점 자리 없애기 :데이터 타입 int로 바꾸기 -> 다시 정의 :기존데이터에 입력
sales_df_2["총이윤"]=sales_df_2["총이윤"].astype(int)
sales_df_2.head()
```

```python
#뒤의 소수점 자리 반올림: round -> 소수점 없애기 int
#sales_df_2["총비용"].astype(int) ->소수점 뒷자리 버림
sales_df_2["총비용"]=sales_df_2["총비용"].round(0).astype(int)
sales_df_2.head()
```

```python
#반올림 한자리 소수점 한자리 
sales_df_2["원가"]=sales_df_2["원가"].round(1)
sales_df_2.head()
```

```python
#round에 음수값 넣어보기 소수점의 갯수 까지 살려두는 것 round(1)1첫쨰짜리 2쨰자리를 반올림해서 앞에숫자 증가,
#round(0)소수점 1 쨰자리 확인후 1의자리로 반올림
sales_df_2["판매개수"]=sales_df_2["판매개수"].round(-1)
sales_df_2.head()
```

```python
sales_df_2["주문날짜"]=sales_df_2["주문날짜"] =pd.to_datetime(sales_df_2["주문날짜"],format="%m/%d/%Y")
sales_df_2.head()


#데이터 형태가 월 /일 /년 이라 그에 맞춰 포맷을 바꿔야함
```

```python
# 데이터 묶기 (묶을것,보여줄것)
#국가별로 묶고 이윤을 다더해줌

sales_df_2.groupby("국가").sum()["총이윤"]

```

```python
#지역에 맞는 국가들의 이윤합 (2가지): 리스트처럼 괄호로 순서대로 묶기
sales_df_2.groupby(["지역","국가","상품 타입"]).sum()["총이윤"]
```

```python
#mean :평균값 나옴
sales_df_2.groupby(["지역","국가","상품 타입"]).mean()["총이윤"]
```

```python
#그룹별로 묶어서 한 국가만 보고싶다거나 한 지역만 보고싶은 경우 get
#묶은 국가의 갯수를 만들고 싶을 경우??? 질문
sales_df_2.groupby("국가").get_group("South Korea")
```

```python
#두지역을 보고싶은 경우???
sales_df_2.groupby("국가").get_group("South Korea")
```

```python
#get group에 여러나라 넣기  ??????????
sales_df_3=sales_df_2[sales_df_2["국가"].isin(["South Korea","China","Togo"])]
sales_df_3.groupby("국가").mean()["총이윤"]
```
