```python
#등록일시 기준 정렬 바꾸기
corona_df_2=corona_df.sort_values("등록일시") #오름차순
corona_df_2
```

```python
corona_df_3=corona_df.sort_values("등록일시",ascending=False) #내림차순
corona_df_3
```

```python
#사망자,확진자 모두 누적-> 1일 확진자, 1일 사망자 확인하기
#1일 확진자 shift(1) 아래로 한칸=기본값 shift()  
#새로운 컬럼 정하기
corona_df_2["일일확진자"]=(corona_df_2['확진자']-corona_df_2['확진자'].shift()).fillna(0)
corona_df_2
```

```python
#일일 사망자
#오름차순
#1일 확진자 shift(1) 아래로 한칸=기본값 shift()  
# shift(n) 아래로 n칸 
#새로운 컬럼 정하기
corona_df_2["일일사망자"]=(corona_df_2['사망자']-corona_df_2['사망자'].shift()).fillna(0)
corona_df_2
```

```python
#diff 아래칸과의 차이 값 
#fillna 결측값을 괄호안 값으로 바꿔줌 filla(a)는 NaN값을 a값으로 바꿈
corona_df_2["일일사망자"]=corona_df_2['사망자'].diff().fillna(0)
corona_df_2
```

```python
#문자열을 데이터 타입으로 변경하는 법
corona_df_6["등록일시"]=pd.to_datetime(corona_df_6["등록일시"], #날짜형으로 변환 0은 월요일 6은 일요일
                                    format= "%Y-%m-%d")
corona_df_6["요일"]=corona_df_6["등록일시"].dt.weekday
corona_df_6.head()
```

```python
#요일에 월화수목금토일 넣기

def week_change(x):
    if x == 0:
        return "월"
    elif x == 1:
        return "화"
    elif x == 2:
        return "수"
    elif x == 3:
        return "목"
    elif x == 4:
        return "금"
    elif x == 5:
        return "토"
    else: #elif x==6: return "일" else: return "" 으로 써줘야함
        return "일"
   
corona_df_6["요일"]= corona_df_6["요일"].apply(week_change)
corona_df_6.head()
```

```python
def week_change_2(x):
    week = ["mon","tue","wed","thu","fri","sat","sun"]
    return week[x]
corona_df_6["요일"]= corona_df_6["요일"].apply(week_change_2)
corona_df_6.head()
```

```python
d = ["월", "화", "수", "목", "금", "토", "일"]  
corona_df_6["요일"] = corona_df_6["요일"].apply(lambda x : d[x])
corona_df_6.head()
```
